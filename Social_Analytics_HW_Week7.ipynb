```python
# Dependencies
import tweepy
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import style
from matplotlib import pylab
style.use('ggplot')

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
#Please get out of NOTES... 


# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
CNN_target_user = "@CNN"
counter = 1
# Variables for holding sentiments
CNN_sentiments = []

# Variable for max_id
CNN_oldest_tweet = None

# Loop through 5 pages of tweets (total 100 tweets)
for x in range(5):

    # Get all tweets from home feed
    CNN_public_tweets = api.user_timeline(CNN_target_user, max_id = CNN_oldest_tweet)

    # Loop through all tweets 
    for CNN_tweet in CNN_public_tweets:  
        # Run Vader Analysis on each tweet
        results = analyzer.polarity_scores(CNN_tweet["text"])
        compound = results["compound"]
        pos = results["pos"]
        neu = results["neu"]
        neg = results["neg"]
        tweets_ago = counter
        
        # Get Tweet ID, subtract 1, and assign to oldest_tweet
        CNN_oldest_tweet = CNN_tweet['id'] - 1
        
        # Add sentiments for each tweet into a list
        CNN_sentiments.append({"Date": CNN_tweet["created_at"], 
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,
                           "Tweets Ago": counter})
        
        # Add to counter 
        counter += 1
```

```python
# Convert sentiments to DataFrame
CNN_sentiments_pd = pd.DataFrame.from_dict(CNN_sentiments)
CNN_sentiments_pd.to_csv("CNN_Sentiments.csv", index=False, header=True)
CNN_sentiments_pd.head()
```
